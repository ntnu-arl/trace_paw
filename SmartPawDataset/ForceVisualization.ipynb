```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set up the file paths to our data set 
filepath = 'E:/Bruker/Dokumenter/Skole/Master/smart_paws/data/machined/'
outputpath = 'E:/Bruker/Dokumenter/Skole/Master/smart_paws/SmartPawDataset/figures/forceEst/'
filepathAruco = filepath + 'Oakd_lite/'
filepathNiclaVision = filepath + 'NiclaVision/'
```

```python
# Test readback
df = pd.read_csv(filepath + 'metadata_normalized.csv')

# df = pd.read_csv(filepath + 'metadata_2.csv')
print(df.head())
```

## Plotting

```python
# Create subplots with 2 rows and 3 columns
fig, axs = plt.subplots(2, 3, figsize=(12, 6))

# Extract the columns from the DataFrame
Fx = df['Fx']
Fy = df['Fy']
Fz = df['Fz']
yaw = df['yaw']
pitch = df['pitch']
roll = df['roll']

# Define slightly dimmed colors
dim_red = 'tab:red'
dim_green = 'tab:green'
dim_blue = 'tab:blue'

# Plot each column in its respective subplot with slightly dimmed colors
axs[0, 0].plot(Fx, color=dim_red)
axs[0, 0].set_title('Fx', fontsize = 20)
axs[0, 0].set_ylabel('Force [N]', fontsize = 12)
axs[0, 1].plot(Fy, color=dim_green)
axs[0, 1].set_title('Fy', fontsize = 20)
axs[0, 1].set_ylabel('Force [N]', fontsize = 12)
axs[0, 2].plot(Fz, color=dim_blue)
axs[0, 2].set_title('Fz', fontsize = 20)
axs[0, 2].set_ylabel('Force [N]', fontsize = 12)
axs[1, 0].plot(pitch, color=dim_red)
axs[1, 0].set_title('Pitch', fontsize = 20)
axs[1, 0].set_ylabel('Angle [Rads]', fontsize = 12)
axs[1, 1].plot(roll, color=dim_green)
axs[1, 1].set_title('Roll', fontsize = 20)
axs[1, 1].set_ylabel('Angle [Rads]', fontsize = 12)
axs[1, 2].plot(yaw, color=dim_blue)
axs[1, 2].set_title('Yaw', fontsize = 20)
axs[1, 2].set_ylabel('Angle [Rads]', fontsize = 12)

# Adjust the spacing between subplots
plt.tight_layout()

# plt.savefig(outputpath + 'forceAngleVis.svg')
# Show the plot
plt.show()
```

```python
# Create subplots with 2 rows and 3 columns
fig, axs = plt.subplots(2, 3, figsize=(12, 6))

# Extract the columns from the DataFrame
Fx = df['Fx']
Fy = df['Fy']
Fz = df['Fz']
yaw = df['yaw']
pitch = df['pitch']
roll = df['roll']

# Define slightly dimmed colors
dim_red = 'tab:red'
dim_green = 'tab:green'
dim_blue = 'tab:blue'

# Plot each column in its respective subplot with slightly dimmed colors
axs[0, 0].hist(Fx, color=dim_red, bins=20)
axs[0, 0].set_title('Fx', fontsize = 20)
axs[0, 0].set_xlabel('Force [N]', fontsize = 12)
axs[0, 1].hist(Fy, color=dim_green, bins=20)
axs[0, 1].set_title('Fy', fontsize = 20)
axs[0, 1].set_xlabel('Force [N]', fontsize = 12)
axs[0, 2].hist(Fz, color=dim_blue, bins=20)
axs[0, 2].set_title('Fz', fontsize = 20)
axs[0, 2].set_xlabel('Force [N]', fontsize = 12)
axs[1, 0].hist(pitch, color=dim_red, bins=20)
axs[1, 0].set_title('Pitch', fontsize = 20)
axs[1, 0].set_xlabel('Angle [Rads]', fontsize = 12)
axs[1, 1].hist(roll, color=dim_green, bins=20)
axs[1, 1].set_title('Roll', fontsize = 20)
axs[1, 1].set_xlabel('Angle [Rads]', fontsize = 12)
axs[1, 2].hist(yaw, color=dim_blue, bins=20)
axs[1, 2].set_title('Yaw', fontsize = 20)
axs[1, 2].set_xlabel('Angle [Rads]', fontsize = 12)

# Adjust the spacing between subplots
plt.tight_layout()

# plt.savefig(outputpath + 'forceAngleHistogram.svg')
# Show the plot
plt.show()
```
